```python
imię = input("Imię:")
if imię.isdigit():
  print ("Niepoprawna wartość")
else:
  print ("Poprawna wartość")
nazwisko = input("Nazwisko:")
if nazwisko.isdigit():
  print ("Niepoprawna wartość")
else:
  print ("Poprawna wartość")
wiek = int(input("Wiek:"))
hasło = input("Hasło:")
powtórz_hasło = input("Powtórz_hasło:")
if hasło == powtórz_hasło:
  print ("Hasła są zgodne")
else:
  print ("Hasła nie są zgodne")
print(("Imię:", imię, "Nazwisko:", nazwisko, "Wiek:", wiek, "Hasło:", hasło))
```

```python
x = int(input("Wprowadź_pierwszą_liczbę:"))
y = int(input("Wprowadź_drugą_liczbę:"))
z = input("Podaj symbol matematyczny(+, -, *, /, %, **)")
if z == "+":
   print(x + y)
elif z == "-":
  print(x - y)
elif z == "*":
  print(x * y)
elif z == "/":
  print(x / y)
elif z == "%":
  print(x % y)
elif z == "**":
  print(x ** y)
else:
  print("Błąd: operator nie znaleziony")
```
