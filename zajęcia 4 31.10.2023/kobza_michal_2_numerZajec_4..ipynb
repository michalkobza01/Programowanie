```python
def zamien_dwucyfrowe_na_string(lista):
    for liczby in range(len(lista)):
        if 10 <= lista[liczby] <= 99:
            lista[liczby] = str(lista[liczby])

zamien_dwucyfrowe_na_string(lista)

print(lista)
```
